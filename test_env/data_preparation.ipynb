```python
import pandas as pd
```

```python
file = pd.read_excel('examples.xlsx', sheet_name="Test 1")
clean_file = file

# Only keep the "Question", "Answer from database", "Answer from user AI" and "Rating" columns
clean_file = clean_file[["Question", "Answer from database", "Answer from AI", "Comment for rating", "Rating"]]
display(clean_file.head())

clean_file = clean_file[clean_file["Rating"].isna() == False]
clean_file["Rating"] = clean_file["Rating"].apply(lambda x: float(x.split("/")[0]))

# Renumber the index
clean_file = clean_file.reset_index(drop=True)

# create id for each row
clean_file["id"] = [i for i in range(1, len(clean_file) + 1)]

display(clean_file)


few_show_rows = clean_file.loc[[1,2,4], ].copy()
display(few_show_rows)

remaining_rows = clean_file.drop([1,2,4], axis=0)

# choose randomly 80% of the rows
train = remaining_rows.sample(frac=0.8, random_state=42).copy()
test = remaining_rows.drop(train.index).copy()

display(train)
display(test)

# Save the train and test files as csv
train.to_csv("train_examples.csv", index=False)
test.to_csv("test_examples.csv", index=False)
few_show_rows.to_csv("few_show_rows.csv", index=False)


# Save the cleaned file as csv
clean_file.to_csv("cleaned_examples.csv", index=False)
```
